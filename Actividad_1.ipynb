## Ejercicio 1
Un estudiante realiza cuatro exámenes en el semestre, los cuales tienen las siguientes
ponderaciones:
o Examen 1, 40%
o Examen 2, 20%
o Examen 3, 10%
o Examen 4, 30%

```python
#Desde la variable Ex1 a Ex4 se pregunta la calificación obtenida en el examen.
#Se utiliza el tipo float para capturar los decimales de cada calificación
#Por último, se asigna una variable donde se calcula la calificación final del alumno (suma) tomando en cuenta la ponderación según el examen; 40%, 20%, 10%, 30%
Ex1=float(input("Ingresa la calificación del examen 1:"))
Ex2=float(input("Ingresa la calificación del examen 2:"))
Ex3=float(input("Ingresa la calificación del examen 3:"))
Ex4=float(input("Ingresa la calificación del examen 4:"))
CFinal=float((Ex1*0.40)+(Ex2*0.20)+(Ex3*0.10)+(Ex4*0.30))
print("La calificación final es de: ", CFinal)
```
